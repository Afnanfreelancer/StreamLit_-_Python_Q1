```python
%%writefile Python_Website.py
import pandas
import streamlit as st
import io
# streamlit.title("Data DashBoard")
st.markdown("<h1 style='color: blue;'>Data Dashboard</h1>", unsafe_allow_html=True)
uploaded_file =st.file_uploader("Chose a csv file", type = 'csv')

if uploaded_file is not None:
  #st.write("File Uploaded ")
  df = pandas.read_csv(uploaded_file)
  st.markdown("<h1 style='color:blue:'>Data Preview</h1>", unsafe_allow_html=True)
  st.write(df.head())
  st.subheader("Description")
  # buffer = io.StringIO()
  # df.info(buf=buffer)
  # info_str = buffer.getvalue()

  # st.text(info_str)
  st.write(df.describe())



  # st.subheader("Filter Data")
  # columns= df.columns.to_list()
  # selected_columns= st.multiselect("Select Columns", columns)
  # new_df = df[selected_columns]
  # st.dataframe(new_df)

  st.subheader("Filter Data")
  columns= df.columns.to_list()
  selected_columns=st.selectbox("Select Columns", columns)
  ## unnique value from selected columns
  unique_values=df[selected_columns].unique()
  selected_values=st.selectbox("Select Values",unique_values)


  filtered_dataframe=df[df[selected_columns]==selected_values]
  st.write(filtered_dataframe)

  st.subheader("Ploting.......")
  x_columns = st.selectbox("Select X_axis Column",columns)
  y_columns = st.selectbox("Select Y_axis Column",columns)

  if st.button("Generate Plot"):
    st.line_chart(filtered_dataframe.set_index(x_columns)[y_columns])
else:
  print("Waiting on File Uploading")

```
