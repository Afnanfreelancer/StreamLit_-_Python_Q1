```python
import random
count = 3
User_Provided_number = int(input("Enter Number in range 1 to 10: "))

while count > 0:
  computer_guess = random.randint(1,10)
  count-=1
  if computer_guess == User_Provided_number:
    print("Winner Winner Chicken Dinner! ")
    print(f"User Number was {User_Provided_number} Computer  Number {computer_guess} ")
    break
  elif computer_guess >= User_Provided_number:
    print(f"\nComputer tried with {computer_guess}")
    print("ooooooh! You were Close\n try something higher Boi \n")

  elif computer_guess <= User_Provided_number:
    print(f"\nComputer tried with {computer_guess}")
    print("ooooooh! You were Close\n try something Lesser Boi \n")

  if count == 0:
    print(f"\nComputer  ran out of tries, actual User number was {User_Provided_number}")
```
