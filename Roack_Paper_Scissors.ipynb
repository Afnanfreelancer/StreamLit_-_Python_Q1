```python
import random
```

```python
def game():

  user= str(input("Enter Choice r for Rock p for paper s for scissor: "))
  pc = random.choice(['r', 's', 'p'])
  if user == pc:
    print("It's a tie! Both chose", user)
  elif user == 'r' and pc == 's':
      print("User wins! User chose Rock and PC chose Scissors")
  elif user == 'r' and pc == 'p':
      print("PC wins! User chose Rock and PC chose Paper")
  elif user == 's' and pc == 'r':
      print("PC wins! User chose Scissors and PC chose Rock")
  elif user == 's' and pc == 'p':
      print("User wins! User chose Scissors and PC chose Paper")
  elif user == 'p' and pc == 'r':
      print("User wins! User chose Paper and PC chose Rock")
  elif user == 'p' and pc == 's':
      print("PC wins! User chose Paper and PC chose Scissors")

game()
```
