```python
import random
count = 3
generaeted_number = random.randint(1,10)
while count > 0:
  count -= 1
  user_input= int(input(f"Enter Number in range 1 to 10 you have 3 tries and  {count}  are left ;) "))

  if user_input == generaeted_number:
    print("Winner Winner Chicken Dinner! ")
    print(f"Generated Number {generaeted_number} Your Number {user_input} ")
    break
  elif user_input >= generaeted_number:
    print("ooooooh! You were Close\n try something higher Boi")

  elif user_input <= generaeted_number:
    print("ooooooh! You were Close\n try something Lesser Boi")

  if count == 0:
    print(f"You ran out of tries, actual number was {generaeted_number}")
```
