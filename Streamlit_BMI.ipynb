```python
%%writefile app.py
import streamlit as st

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi

def main():
    st.title("BMI Calculator")

    weight = st.slider("Select your weight (kg)", 30, 150, 70)
    height = st.slider("Select your height (m)", 1.2, 2.2, 1.7)

    if st.button("Calculate BMI"):
        bmi = calculate_bmi(weight, height)
        st.write(f"Your BMI is: {bmi:.2f}")

        if bmi < 18.5:
            st.write("You are underweight.")
        elif 18.5 <= bmi < 24.9:
            st.write("You have a normal weight.")
        elif 25 <= bmi < 29.9:
            st.write("You are overweight.")
        else:
            st.write("You are obese.")
main()

#Steps to Run in Google Colab:
#1. Install dependencies:
#!pip install streamlit pyngrok
#2. Save the above code in a file:
 #  Run this cell as it is already using %%writefile to save the script.
#3. Run Streamlit and expose it using LocalTunnel:

```
